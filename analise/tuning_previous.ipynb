# importacoes

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython.display import display
```

```python
def printa_df(df):
    
    print(df.shape)
    display(df.head())
    
pd.set_option('display.max_columns', None)
```

```python
df_aplicacao = pd.read_csv('../../../../base_de_dados/previous_application.csv')

printa_df(df_aplicacao)
```

```python
target = pd.read_csv(r'../../../../base_de_dados/application_train.csv')[['SK_ID_CURR', 'TARGET']]

target.set_index('SK_ID_CURR', inplace=True)

printa_df(target)
```

# analise

```python
df_aplicacao.replace('XNA', np.nan, inplace=True)

printa_df(df_aplicacao)
```

```python
nulos = pd.DataFrame(df_aplicacao.isnull().sum(), columns=['Nulos'])
nulos = nulos[nulos['Nulos'] >= df_aplicacao.shape[0] * 0.3]
df_aplicacao.drop(nulos.index, axis=1, inplace=True)
```

```python
nulos
```

```python
df_analise = df_aplicacao.join(target, how='inner')

printa_df(df_analise)
```

```python
features_cat = df_analise.select_dtypes('O').columns

for feature in features_cat:
    
    display(df_analise.groupby(feature).agg({'TARGET':['mean', 'count']}))
```

```python
df_aplicacao.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

for col in features_cat:
    df_aplicacao[col] = le.fit_transform(df_aplicacao[col].astype(str))
    for i, v in enumerate(le.classes_):
        if v == 'nan':
            df_aplicacao[col].replace(i, np.nan,inplace=True)
```

```python
from missingpy import MissForest
```

```python
x = df_aplicacao.drop('SK_ID_PREV', axis=1)
mf = MissForest(max_depth=10, max_iter=3, n_estimators=10, max_features=0.7, min_samples_split=10,
                min_impurity_decrease=0.01)

mf.fit(x)
```

```python
transform = mf.transform(x)
```

```python
df_aplicacao[x.columns] = transform
```

```python
df_aplicacao['valor_solicitado/liberado'] = df_aplicacao['AMT_APPLICATION'] - df_aplicacao['AMT_CREDIT']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first', sparse=False)
for col in features_cat:
    ohe.fit(df_aplicacao[[col]])
    labels = [col + '_' + str(i + 0) for i in range(len(ohe.categories_[0]) - 1)]
    
    df_aplicacao[labels] = ohe.transform(df_aplicacao[[col]])
```

```python
df_saida = df_aplicacao.groupby('SK_ID_CURR').mean()

df_saida.reset_index(inplace=True)

printa_df(df_saida)
```

```python
df_saida.to_csv(r'../../../../base_de_dados/features_aplicacao.csv', sep=';', index=False)
```
